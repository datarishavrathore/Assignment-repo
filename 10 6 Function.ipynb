# 5 Types of Params
1. Positional only
2. Keyword only
3. Positional or keyword
4. Variable length Positional(*a)
5. Variable length Keyword(**a)

```python
# Positional only
def show(a,b,/): # forward slash se only positiona agrs hi valid honge keyword args se error ayegi
    print(a,b)
show (10,20)
# error show(a=4,b=8) # error keyword args pe error
```

```python
# Keyword only (* asterik lgane se keyword ayenge only)
def show (*,a,b):  # a,b are keyword only params params se phele * lgane se yeh only keyword args hi accept krega position pe error 
    print(a,b)
show(a=4,b=4)
# show(4,b=5) # error
```

```python
# Positional or keyword
####  Note position phele keyword baad mein
#a,b----->pos only
# e,f---->keyword only
# c,d---->pos or keyword 

def show(a,b,/,c,d,*,e,f):
    print(a,b,c,d,e,f)

show(1,2,3,d=4,e=9,f=4)
# show(1,2,c=3,4,e=5,f=6)   # error becuase first positional aata tha then keyword aata hai
```

```python
# Variable length Positional (*a)   Arbitrary Arguments (*args)
# asterik k baad comma nhi lgana hai 
def show(*a):# work as tuple  #asterik lgane se yeh kitne bhi postional de skte hai koi limit nhi hai
    print(a)

show(10,20,30,50,60)
show(1,34,67,46,7)
show(1)
```

```python
# Variable length Keyword(**a) (**kwargs)
# double asterik lgte hai 


def show(**a):   # varible length keyword(works as dict)
    print(a)

show(roll=101,name='sonu',age=45)
show(a=10,ename='text')
show()
```

```python
# return // a function can also retrun it's local data to calling stms
# we can use only 1 return stms
# if we want to retrun multiple values the we can use list/tuple
# if we do not return any value/do not write retrun then none is default retrun value

# return k neeche likha hua code nhi run hoga
# retrun k baad function stop
# default return value is none
def mul(a,b):
    c=a*b
    return c
res=mul(4,5)
print(res)

# None aaya kyuki ismein retrun likhte nhi hai return nhi hai
def mul(a,b):
    c=a*b
res=mul(3,4)
print(res)
```

```python
# ek function se 2 value return ho skte hai lekin group bna k means TUPLE bna ke
def calc(a,b):
    c=a+b
    d=a*b
    return (c,d)
    
res=calc(3,4)
print(res)
```

```python
# call krna 
res=calc(7,3)
print(res)
```

```python
# direct way to call
print(calc(2,3))
```

```python
# wap to check string is palindrome or not palidrome
def check_palindrome():
    name = input("Enter the name: ").lower()
    if name == name[::-1]:   # reverse
        print("This is palindrome")
    else:
        print("Not palindrome")
check_palindrome()
```

```python
def hello(password,username):
    print("value of pass :",password)
    print("value of user :",username)
hello(username="username",password="password")
```

```python
# keyword(**kwargs)
def hello(**kwargs):
    print("value of pass:", kwargs.get('password'))
    print("value of user:", kwargs.get('username'))
    print("value of email:", kwargs.get('email'))

hello(username="aman-user", password="password@1234", bio="")
```
