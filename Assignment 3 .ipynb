```python
#Write a Python program to check if a number provided by the user is prime or not.
num = int(input("Enter a number: "))
if num <= 1:
    print("Not Prime")
else:
    prime = True
    for i in range(2, num):
        if num % i == 0:
            prime = False
            break
    if prime:
        print("Prime Number")
    else:
        print("Not Prime Number")
```

```python
#Write a Python program to generate the Fibonacci sequence up to a specified number of terms.
terms = int(input("Enter number of terms: "))
a = 0
b = 1
for i in range(terms):
    print(a)
    c = a + b
    a = b
    b = c
```

```python
#Write a Python program that allows the user to search for a character within a given string. 
s = input("Enter a string: ")
ch = input("Enter a character to search: ")

found = False

for i in s:
    if i == ch:
        found = True
        break

if found:
    print("Character found")
else:
    print("Character not found")
```

```python
#Write a Python program to filter out duplicate characters from a string entered by the user
a = input("Enter the string: ")

dup = ""

for i in a:
    if a.count(i) > 1 and i not in dup:
        dup += i

print("Duplicate characters:", dup)
```

```python
# Take string from user and Replace every vowel in the string with an asterisk *
text = input("Enter a string: ")

result = ""

for i in text:
    if i.lower() in "aeiou":
        result += "*"
    else:
        result += i

print("Modified string:", result)
```
