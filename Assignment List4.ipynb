```python
#Write a Python program to display the odd and even elements from the list [10,23, 11, 12, 33, 44, 2, 5, 6]
var=[10,23, 11, 12, 33, 44, 2, 5, 6]
def odd(var):
    for i in var:
        if i%2==0:
            print("even")
        else:
            print("odd")
odd(var)
```

```python
#Write a Python program to count the odd and even numbers in the list [10, 23,11, 12, 33, 44, 2, 5, 6].
var=[10, 23,11, 12, 33, 44, 2, 5, 6]
count=0
for i in var:
    if i%2==0:
        print("even")
    else:
        print("odd")
    count+=1
    print(count)
```
